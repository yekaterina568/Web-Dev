Task A

```python
def min4(a, b, c, d):
    res = a
    if b < res: res = b
    if c < res: res = c
    if d < res: res = d
    return res

vals = list(map(int, input().split()))
print(min4(vals[0], vals[1], vals[2], vals[3]))
```

Task B

```python
def power(a, n):
    res = 1.0
    for i in range(n):
        res *= a
    return res

vals = input().split()
a = float(vals[0])
n = int(vals[1])

print(power(a, n))
```

Task C

```python
def xor(x, y):
    if x != y:
        return True
    else:
        return False

v = list(map(int, input().split()))
x = bool(v[0])
y = bool(v[1])

if xor(x, y):
    print(1)
else:
    print(0)
```
