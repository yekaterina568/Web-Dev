Task A

```python
n = int(input())
a = list(map(int, input().split()))

for i in range(0, n, 2):
    print(a[i], end=' ')
```

Task C

```python
n = int(input())
a = list(map(int, input().split()))
c = 0

for x in a:
    if x > 0:
        c += 1

print(c)
```

Task D

```python
n = int(input())
a = list(map(int, input().split()))
c = 0

for i in range(1, n):
    if a[i] > a[i - 1]:
        c += 1

print(c)
```

Task E

```python
n = int(input())
a = list(map(int, input().split()))
found = False

for i in range(1, n):
    if a[i] * a[i - 1] > 0:
        found = True
        break

print("YES" if found else "NO")
```

Task F

```python
n = int(input())
a = list(map(int, input().split()))
c = 0

for i in range(1, n - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        c += 1

print(c)
```

Task G

```python
n = int(input())
a = list(map(int, input().split()))

for i in range(n // 2):
    a[i], a[n - 1 - i] = a[n - 1 - i], a[i]

print(*(a))
```
