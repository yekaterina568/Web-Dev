count_evens

```python
def count_evens(nums):
    count = 0
    for n in nums:
        if n % 2 == 0:
            count += 1
    return count
```

big_diff

```python
def big_diff(nums):
    return max(nums) - min(nums)
```

centered_average

```python
def centered_average(nums):
    res = (sum(nums) - max(nums) - min(nums)) // (len(nums) - 2)
    return res
```

sum13

```python
def sum13(nums):
    total = 0
    i = 0
    while i < len(nums):
        if nums[i] == 13:
            i += 2 
            continue
        total += nums[i]
        i += 1
    return total
```

sum67

```python
def sum67(nums):
    total = 0
    skip = False
    for n in nums:
        if n == 6:
            skip = True
        if not skip:
            total += n
        if n == 7 and skip:
            skip = False
    return total
```

has22

```python
def has22(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 2 and nums[i+1] == 2:
            return True
    return False
```
