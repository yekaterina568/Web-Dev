first_last6

```python
def first_last6(nums):
    if nums[0] == 6 or nums[-1] == 6:
        return True
    else:
        return False
```

common_end

```python
def common_end(a, b):
    if a[0] == b[0] or a[-1] == b[-1]:
        return True
    else:
        return False
```

reverse3

```python
def reverse3(nums):
    return [nums[2], nums[1], nums[0]]
```

middle_way

```python
def middle_way(a, b):
    return [a[1], b[1]]
```

same_first_last

```python
def same_first_last(nums):
    return len(nums) >= 1 and nums[0] == nums[-1]
```

sum3

```python
def sum3(nums):
    return sum(nums)
```

max_end3

```python
def max_end3(nums):
    m = max(nums[0], nums[-1])
    return [m, m, m]
```

make_ends

```python
def make_ends(nums):
    return [nums[0], nums[-1]]
```

make_pi

```python
def make_pi():
    return [3, 1, 4]
```

rotate_left3

```python
def rotate_left3(nums):
    return [nums[1], nums[2], nums[0]]
```

sum2

```python
def sum2(nums):
    if len(nums) == 0:
        return 0
    return sum(nums[:2])
```

has23

```python
def has23(nums):
    if 2 in nums or 3 in nums:
        return True
    else:
        return False
```
