hello_name

```python
def hello_name(name):
    return "Hello " + name + "!"
```

make_out_word

```python
def make_out_word(out, word):
    return out[:2] + word + out[2:]
```

first_half

```python
def first_half(s):
    mid = len(s) // 2
    return s[:mid]
```

non_start

```python
def non_start(a, b):
    return a[1:] + b[1:]
```

make_abba

```python
def make_abba(a, b):
    return a + b + b + a
```

extra_end

```python
def extra_end(s):
    last_two = s[-2:]
    return last_two * 3
```

without_end

```python
def without_end(s):
    return s[1:-1]
```

left2

```python
def left2(s):
    return s[2:] + s[:2]
```

make_tags

```python
def make_tags(tag, word):
    return "<" + tag + ">" + word + "</" + tag + ">"
```

first_two

```python
def first_two(s):
    return s[:2]
```

combo_string

```python
def combo_string(a, b):
    if len(a) < len(b):
        return a + b + a
    else:
        return b + a + b
```
