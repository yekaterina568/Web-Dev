1. Python If-Else

```python
n = int(input())
if n % 2 != 0:
    print("Weird")
else:
    if n >= 2 and n <= 5:
        print("Not Weird")
    elif n >= 6 and n <= 20:
        print("Weird")
    else:
        print("Not Weird")
```

2. Find the Runner-Up Score!

```python
n = int(input())
nums = list(map(int, input().split()))

first = -101
second = -101

for x in nums:
    if x > first:
        second = first
        first = x
    elif x > second and x != first:
        second = x

print(second)
```

3. Nested Lists

```python
students = []
for i in range(int(input())):
    name = input()
    score = float(input())
    students.append([name, score])

scores = sorted(list(set([x[1] for x in students])))
second_low = scores[1]

names = [x[0] for x in students if x[1] == second_low]
names.sort()

for n in names:
    print(n)
```

4. String Split and Join

```python
def split_and_join(line):
    words = line.split(" ")
    res = "-".join(words)
    return res

line = input()
print(split_and_join(line))
```

5. Mutations

```python
def mutation(string, position, char):
    s_list = list(string)
    s_list[position] = char
    return "".join(s_list)

s = input()
p, c = input().split()
print(mutation(s, int(p), c))
```

6. Find a string

```python
def count_substring(string, sub_string):
    count = 0
    for i in range(len(string) - len(sub_string) + 1):
        if string[i : i + len(sub_string)] == sub_string:
            count += 1
    return count

s = input()
sub = input()
print(count_substring(s, sub))
```

7. String Validators

```python
s = input()

for method in ['isalnum', 'isalpha', 'isdigit', 'islower', 'isupper']:
    print(any(getattr(char, method)() for char in s))
```

8. Introduction to Sets

```python
def average(array):
    unique = set(array)
    return sum(unique) / len(unique)

n = int(input())
arr = list(map(int, input().split()))
print(average(arr))
```
